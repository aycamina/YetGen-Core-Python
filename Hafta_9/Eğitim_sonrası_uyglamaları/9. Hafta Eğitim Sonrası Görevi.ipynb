# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
class Kumanda:
    def __init__(self):
        self.tv_on = False
        self.volume = 0
        self.channels = ["TRT", "SHOW", "FOX","KANAL D"]
        self.current_channel = None
    
    def tv_ac(self):
        if not self.tv_on:
            self.tv_on = True
            self.current_channel = self.channels[0]
            print("TV açıldı")
        else:
            print("TV zaten açık")

    def tv_kapat(self):
        if self.tv_on:
            self.tv_on = False
            print("TV kapandı")
        else:
            print("TV zaten kapalı")
    
    def ses_ayarlari(self, volume):
        if self.tv_on:
            self.volume = volume
            print(f"Ses {self.volume} olarak ayarlandı.")
        else:
            print("TV kapalı. Lütfen önce açın.")

    def kanal_ekle(self, channel):
        self.channels.append(channel)
        print(f"Kanal {channel} eklendi")
    
    def kanal_sayisini_ogrenme(self):
        print(f"Kanal sayısı: {len(self.channels)}")
    
    def rastgele_kanala_gecme(self):
        if self.tv_on:
            import random
            self.current_channel = self.channels[random.randint(0, len(self.channels) - 1)]
            print(f"{self.current_channel} kanalına geçildi")
        else:
            print("TV kapalı. Lütfen önce açın.")
    
    def televizyon_bilgileri(self):
        if self.tv_on:
            print(f"TV açık, Mevcut kanal: {self.current_channel}, Ses: {self.volume}, Kanallar: {self.channels}")
        else:
            print("TV kapalı")

kumanda = Kumanda()

def main():
    print("\nSeçenekler:")
    print("1. Tv Aç")
    print("2. Tv Kapat")
    print("3. Ses Ayarları")
    print("4. Kanal Ekle")
    print("5. Kanal Sayısını Öğrenme")
    print("6. Rastgele Kanala Geçme")
    print("7. Televizyon Bilgileri")
    print("q. Çıkış")
    
    while True:
        choice = input("Bir seçenek seçin: ")
        
        if choice == "1":
            kumanda.tv_ac()
        elif choice == "2":
            kumanda.tv_kapat()
        elif choice == "3":
            volume = int(input("Ses seviyesini ayarlayın: "))
            kumanda.ses_ayarlari(volume)
        elif choice == "4":
            channel = input("Kanal ekleyin: ")
            kumanda.kanal_ekle(channel)
        elif choice == "5":
            kumanda.kanal_sayisini_ogrenme()
        elif choice == "6":
            kumanda.rastgele_kanala_gecme()
        elif choice == "7":
            kumanda.televizyon_bilgileri()
        elif choice == "q":
            print("Çıkılıyor...")
            break
        else:
            print("Geçersiz seçenek, lütfen tekrar deneyin.")

if __name__ == "__main__":
    main()

    


```
