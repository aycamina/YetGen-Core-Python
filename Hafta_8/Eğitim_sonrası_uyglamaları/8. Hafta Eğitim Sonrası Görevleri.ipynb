# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
def main():
    giris_sayisi = int(input("Giris sayisi: "))
    for i in range(giris_sayisi):
        isim = input("İsim: ")
        soyisim = input("Soyisim: ")
        yas = input("Yas: ")
        boy = input("Boy: ")
        kilo = input("Kilo: ")
        vki = vki_hesaplama(boy,kilo)
        aralik = araliklar(vki)
        cikti = ciktilar(isim,soyisim,yas,boy,kilo,aralik)

        with open("vki_hesap", "a",) as file:
            file.write(cikti)

def vki_hesaplama(boy,kilo):
    vki = int(kilo)/(int(boy)**2)
    return round(vki,1)
def araliklar(vki):
    if vki<=18.5:
        return "Zayıf"
    elif 18.5<vki<=24.9:
        return "Normal Kilolu"
    elif 24.9<vki<=29.9:
        return "Fazla Kilolu"
    elif 29.9<vki<=39.9:
        return "Obez"
    else:
        return "Morbid obez"
    
def ciktilar(isim,soyisim,yas,boy,kilo,aralik):
    isim = f"İsim: {isim}\n"
    soyisim = f"Soyisim: {soyisim}\n"
    yas = f"Yas: {yas}/n"
    boy = f"Boy: {boy}/n"
    kilo = f"Kilo: {kilo}/n"
    vki = f"VKI: {aralik}/n"

    return isim + soyisim + yas + boy + kilo +aralik

main()
```

**YetBar - Smoothie Bank**

YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
menu = { 
    "elma":0.50,
    "armut":0.75,
    "uzum":2,
    "cilek":1.75,
    "muz":2,
    "kavun":2.15,
    "karpuz":1.99
}
def menu_goster():
    print("--------YetBar Menu--------")
    for item in list(menu.keys()):
        print(f"{item} : {menu[item]} TL")

def tek_urun_fiyat_hesaplama(urun,adet):
    assert adet>=0, "Ürün adedi 0 ve altında olamaz."
    if urun not in list(menu.keys()):
        return None
    else:
        return menu[urun] * adet
    
def satin_alma():
    menu_goster()
    print("Almak istediğiniz ürünü giriniz. Çıkmak içi 0'a basın.")

    alinan_urunler = []
    adet = []
    son_fiyat = 0

    while 1:
        eklenecek_urun = input("Hangi ürünü eklemek istiyorsunuz?")
        if eklenecek_urun =="0":
            print("Fiş kaydediliyor")
            break
        ekleme_adedi = int(input("Kaç adet eklemek istiyorsunuz?"))
        fiyat = tek_urun_fiyat_hesaplama(eklenecek_urun,adet)
        if fiyat == None:
            print("Düzgün değer giriniz.")
        else:
            print(f"Satın alma ücreti: {fiyat}")    
            print("Satın almayı onaylamak için 'y' tuşuna basınız. İptal için 'n' tuşuna basınız.")

        check = input("Satın almayı onaylıyor musunuz?")
        if check == "y":
            print("Satın alma onaylandı")
            print("--------------------")
            son_fiyat+=fiyat
            alinan_urunler.append(eklenecek_urun)
            adet.append(ekleme_adedi)
            print(f"Ara Toplam : {son_fiyat}")
        elif check == "n":
            print("Satın alma işlemi iptal edildi.")


    return fiyat,alinan_urunler,adet

def main():
    print("Yetbar'a hoş geldiniz. Menüyü görüntülemek için 1'e, smoothie yapmak içinlütfen 2'ye basınız.")
    print("Çıkmak için '0' giriniz")

    program= None

    while 1:
        program = int(input("Tercihiniz:"))
        if program<0 or program>2:
            print("Yanlış bir tercih yaptınız. Lütfen tekrar deneyiniz.")
        elif program == 0:
            print("Çıkış yapılıyor.")
        elif program == 1:
            menu_goster()

        elif program==2:
            fiyat,alinan_urunler,adet = satin_alma()

        with open("kasa.txt","a",encoding="utf-8") as f:
            f.write(str(time.ctime(time.time())))
            f.write("\n")
            f.write(','.join(map(str, alinan_urunler)))
            f.write("\n")
            f.write(','.join(map(str, adet)))
            f.write("\n")
            f.write(f"Toplam: {fiyat}")
            f.write("\n")

        cıkıs_kontrol = input("Başka bir işlem yapmay istiyor musunuz? (evet-hayır):")
        if cıkıs_kontrol == "evet":
            pass
        else:
            print("Güle güle")

main()
```
