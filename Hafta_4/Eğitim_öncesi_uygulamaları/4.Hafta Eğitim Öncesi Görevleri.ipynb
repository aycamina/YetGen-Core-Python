# Problem 1
Sözlükleri kullanarak bir telefon rehberi yazın. Bu rehberde kullanıcıya kimin telefonunu görüntülemek istediğini sorun ve kullanıcının girdiği isme göre o kişinin telefon numarasını yazdırın. Proje sonunda elde edeceğiniz çıktı  şuna benzer olmalı:


Lütfen numarasını öğrenmek istediğiniz kişinin adını girin: Ahmet

Ahmet isimli kişinin numarası şu şekildedir: 0532 678 13 19

````python
# 1.Problemin Çözümünü Buraya Yazınız
telefon_numaraları = {"Ahmet": "0532 678 13 19", "Mehmet": "0555 555 55 55", "Fatma": "0544 444 44 44"}
isim = input("Lütfen numarasını öğrenmek istediğiniz kişinin adını girin: ")
if isim in telefon_numaraları:
      numara = telefon_numaraları[isim]
else:
     numara = "Kişi bulunamadı."
print(f"{isim} isimli kişinin numarası şu şekildedir: {numara}")
````

# Problem 2
Sözlükleri kullanarak bir şirket çalışanları indeksi oluşturun. Bu isim indeksinde kişilerin isimleri key, kişilerin  memleket, yaş ve görev bilgileri value olmalıdır. Burada kullanacağımız value değerleri liste olmalıdır.


Daha sonra bir isim sorgulama ekranı gibi kullanıcıya kimin bilgilerini görüntülemek istediğini 
sorun ve sorgulanan kişinin ekranda gösterilmesini sağlayın. Proje sonunda elde edeceğiniz çıktı şu şekilde olmalı:


Lütfen bilgilerini görüntülemek istediğiniz çalışanın ismini girin: Mehmet Yağız

Mehmet Yağız= Memleket: Adana Yaş: 40 Görev: Direktör

````python
# 2.Problemin Çözümünü Buraya Yazınız
çalışanlar = {"Mehmet Yağız":{"Memleket":"Adana","Yaş":"40","Görev":"Direktör"},"Ayşe":{"Memleket":"İstanbul","Yaş":"35","Görev":"Müdür"},"Fatma":"Sivas","Yaş":"55","Görev":"CFO"}
isim = input("Lütfen bilgilerini görüntülemek istediğiniz çalışanın ismini girin: ")
if isim in çalışanlar:
    çalışan_bilgisi = çalışanlar[isim]
    print(f"{isim}= Memleket: {çalışan_bilgisi['Memleket']} Yaş: {çalışan_bilgisi['Yaş']} Görev: {çalışan_bilgisi['Görev']}")
else:
    print("Çalışan bulunamadı.")
````

# Problem 3

3 öğrenciden oluşan bir öğrenci not sözlüğü oluşturun. Bu sözlükte öğrencilerin notları value olarak bir listede toplansın.

Kullanıcıya hangi öğrencinin notlarını görmek istediğini sorun. Öğrencinin notu görüntülendiğinde program sonunda şöyle bir çıktı elde etmelisiniz:

```
Lütfen notlarını görmek istediğiniz öğrencinin adını girin: Mehmet

Mehmet isimli öğrencinin      1.Sınav Notu:72
                              2.Sınav Notu:66
                              3.Sınav Notu:48
Not Ortalaması: 62.0


````python
# 3.Problemin Çözümünü Buraya Yazınız
öğrenciVEnotları = {"Mehmet":{"1. Sınav Notu": 72,"2. Sınav Notu":66,"3. Sınav Notu":48},"Ahmet":{"1. Sınav Notu":45,"2. Sınav Notu":55,"3. Sınav Notu":65},"Ayşe":{"1. Sınav Notu":95,"2. Sınav Notu":100,"3. Sınav Notu":100}}
isim = input("Lütfen notlarını görüntülemek istediğiniz öğrencinin adını girin: ")
if isim in öğrenciVEnotları:
    notlar=öğrenciVEnotları[isim]
    toplam = sum(notlar.values())
    ortalama = toplam / len(notlar)
    print(f"{isim} isimli öğrencinin     1. Sınav Notu: {notlar['1. Sınav Notu']}\n 2. Sınav Notu: {notlar['2. Sınav Notu']}\n 3. Sınav Notu: {notlar['3. Sınav Notu']}")
    print(f"Not Ortalaması: {ortalama:.2f}")
else:
    print("Öğrenci bulunamadı") 

         
````

````python
# 1- "Bmw, Mercedes, Opel, Mazda" elemanlarına sahip bir liste oluşturunuz.
cars = ["Bmw", "Mercedes", "Opel", "Mazda"]

# 2- Liste Kaç elemanlıdır?
list_length = len(cars)
print(f"Liste {list_length} elemanlıdır.")

# 3- Listenin ilk ve son elemanı nedir?
first_car = cars[0]
last_car = cars[-1]
print(f"Listenin ilk elemanı: {first_car}, son elemanı: {last_car}")

# 4- Mazda değerini Toyota ile değiştirin.
cars[cars.index("Mazda")] = "Toyota"
print("Güncellenmiş liste:", cars)

# 5- Listenin -2 indeksindeki değer nedir?
second_last_car = cars[-2]
print(f"Listenin -2 indeksindeki değer: {second_last_car}")

# 6- Listenin ilk 3 elemanını alın.
first_three_cars = cars[:3]
print("Listenin ilk 3 elemanı:", first_three_cars)

# 7- Listenin son 2 elemanı yerine "Toyota" ve "Renault" değerlerini ekleyin.
cars[-2:] = ["Toyota", "Renault"]
print("Son 2 elemanı değiştirilmiş liste:", cars)

# 8- Listenin üzerine "Audi" ve "Nissan" değerlerini ekleyin.
cars.extend(["Audi", "Nissan"])
print("Listeye eklenmiş Audi ve Nissan:", cars)

# 9- Listenin son elemanını silin.
cars.pop()
print("Son elemanı silinmiş liste:", cars)

# 10- Liste elemanlarını tersten yazdırınız.
reversed_cars = cars[::-1]
print("Liste elemanlarının tersten hali:", reversed_cars)

# 11- Aşağıdaki verileri bir liste içinde saklayınız.
studentA = ["Yiğit Bilgi", 2010, (70, 60, 70)]
studentB = ["Sena Turan", 1999, (80, 80, 70)]
studentC = ["Ahmet Turan", 1998, (80, 70, 90)]
students = [studentA, studentB, studentC]

# 12- Liste elemanlarını ekrana yazdırınız.
for student in students:
    print(student)

names = ['Ali', 'Yağmur', 'Hakan', 'Deniz']
years = [1998, 2000, 1998, 1987]

# 13- "Cenk" ismini listenin sonuna ekleyiniz.
names.append("Cenk")
print("Sonuna Cenk eklenmiş liste:", names)

# 14- "Sena" değerini listenin başına ekleyiniz.
names.insert(0, "Sena")
print("Başına Sena eklenmiş liste:", names)

# 15- "Deniz" ismini listeden siliniz.
names.remove("Deniz")
print("Deniz ismi silinmiş liste:", names)

# 16- "Ali" listenin bir elemanı mıdır?
is_ali_in_list = "Ali" in names
print(f"Ali listenin bir elemanı mıdır? {is_ali_in_list}")

# 17- Liste elemanlarını ters çevirin.
names.reverse()
print("Liste elemanlarının ters hali:", names)

# 18- Liste elemanlarını alfabetik olarak sıralayınız.
names.sort()
print("Liste elemanlarının alfabetik sıralanmış hali:", names)

# 19- years listesini rakamsal büyüklüğe göre sıralayınız.
years.sort()
print("Years listesinin sıralanmış hali:", years)

# 20- str = "Chevrolet,Dacia" karakter dizisini listeye çeviriniz.
str_cars = "Chevrolet,Dacia"
car_list_from_str = str_cars.split(',')
print("Karakter dizisinden listeye çevrilmiş hali:", car_list_from_str)

# 21- years dizisinde kaç tane 1998 değeri vardır?
count_1998 = years.count(1998)
print(f"Years dizisinde {count_1998} tane 1998 değeri vardır.")

# 22- years dizisinin tüm elemanlarını siliniz.
years.clear()
print("Years dizisinin tüm elemanları silindikten sonra:", years)

# 23- Kullanıcıdan alacağınız 3 tane marka bilgisini bir listede saklayınız.
user_brands = []
for i in range(3):
    brand = input("Bir marka ismi girin: ")
    user_brands.append(brand)
print("Kullanıcıdan alınan marka listesi:", user_brands)
````
